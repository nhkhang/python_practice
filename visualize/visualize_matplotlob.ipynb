```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Linear data
x1 = [1, 2, 3, 4] 
x2 = [1, 2, 3, 4]
y2 = [4, 2, 0, -2]

plt.plot(x1, x1, label = 'linear 1') # x1 = y1
plt.plot(x2, y2, label = 'linear 2')

plt.legend() # Show description
plt.show()
```

```python
# Drawing data point: plt.scatter()
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='+')
plt.scatter([1.3, 2.1, 3.2, 1.6], [15, 10, 28, 20], color='red', marker='^')

plt.xlim(0.5, 4.5) # Limit of x
plt.show()
```

```python
# Initialize figure
fig = plt.figure()

# Add axes
fig.add_axes([0, 0, 1, 1])
```
